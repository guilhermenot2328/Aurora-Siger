# VARIÁVEIS

```python
import pandas as pd

dados = pd.read_csv("telemetria_dados.csv")
```

```python
# Métricas #

temperaturaInterna = float(dados["temperatura_interna"].mean())
temperaturaExterna = float(dados["temperatura_externa"].mean())
integridadeEstrutural = False
energia = 92
pressaoTanque = int(dados["pressao_tanque"].mean())
modulosCriticos = "Ok"

interrupcao = False
```

# Alternativa 1

```python
def abortar(mensagem):
    print(mensagem)
    return True

condicoes_decolagem = {
    "Temperatura interna (RISCO)": temperaturaInterna < 15 or temperaturaInterna > 35,
    "Temperatura Externa (RISCO)": temperaturaExterna < -20 or temperaturaExterna > 40,
    "Integridade Estrutural (RISCO)": integridadeEstrutural == False,
    "Bateria (RISCO)": energia < 80,
    "Pressão do Tanque (RISCO)": pressaoTanque < 250 or pressaoTanque > 320,
    "Módulos Criíticos (RISCO)": "Ok",
}

for mensagem_erro, condicao in condicoes_decolagem.items():
    if condicao == True:
        interrupcao = abortar(mensagem_erro)

if interrupcao == True:
    print("="*35, "\nMissão abortada!")
else:
    print("Pronto para a decolagem")
```

# Alternativa 2

```python
def abortar(mensagem):
    print(mensagem)
    return True

if temperaturaInterna < 15 or temperaturaInterna > 35:
    abortamento = abortar("Temperatura interna (RISCO)")

if temperaturaExterna < -20 or temperaturaExterna > 40:
    abortamento = abortar("Temperatura Externa (RISCO)")

if integridadeEstrutural == False:
    abortamento = abortar("Integridade Estrutural (RISCO)")

if energia < 80:
    abortamento = abortar("Bateria (RISCO)")

if pressaoTanque < 250 or pressaoTanque > 320:
    abortamento = abortar("Pressão do Tanque (RISCO)")

if modulosCriticos != "Ok":
    abortamento = abortar("Módulos Criíticos (RISCO)")
    
if abortamento == True:
    print("="*35, "\nMissão abortada!")
else:
    print("Pronto para a decolagem")
```
